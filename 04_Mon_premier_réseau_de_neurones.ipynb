Jusqu'à présent, nous avons vu comment créer des modèles de Régression Logistique.  
Ces modèles sont idéaux pour résoudres de problèmes simples, tels que la séparation linéaire de deux groupes de points.  
Cependant, avec des problèmes plus sophistiqués, le caractère linéaire de nos modèles le rendent biaisé, car ce dernier ne peut pas comprendre la complexité du problème.

En Machine Learning, il est intéressant de rajouter des variables à certaines puissances (tels qu'un x1^2) en entrée, pour développer un modèle polynomial.  
On appelle cela du "feature engineering".  
Ce travail est cependant lent et demande une certaine expertise (que je n'ai bien évidemment pas actuellement).
De plus, nous voulons faire du Deep Learning, non pas du Machine Learning.

Nous allons donc améliorer notre modèle en lui rajoutant des neurones.
L'objectif est de laisser notre modèle fabriquer tout seul son propre "feature engineering" en lui allouant des neurones dédiés à cela.


Cela nous permettra d'avoir un modèle non linéaire et qui peut apprendre à résoudres des problèmes beaucoup plus complexes.
